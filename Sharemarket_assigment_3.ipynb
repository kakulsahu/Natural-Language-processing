```python
import spacy
import pandas as pd
from collections import Counter
```

```python
# Load spaCy model
nlp = spacy.load("en_core_web_sm")
share_market_paragraph = """
The share market, also known as the stock market, is a platform where investors buy and sell shares of publicly listed companies.
Major stock exchanges include the New York Stock Exchange, NASDAQ, Bombay Stock Exchange, and National Stock Exchange.
Companies such as Apple, Microsoft, Google, Reliance Industries, and Tata Consultancy Services are actively traded.
Investors participate in the share market through brokers like Zerodha, Upstox, Robinhood, and Charles Schwab.
Stock prices are influenced by economic indicators, company performance, interest rates, and global events.
The Securities and Exchange Board of India regulates the Indian share market.
In the United States, the Securities and Exchange Commission oversees stock market operations.
Share market indices such as Sensex, Nifty 50, Dow Jones, and S&P 500 track market performance.
Initial Public Offerings allow companies to raise capital from the public.
Investors use strategies like long-term investing, day trading, and swing trading.
Mutual funds and exchange-traded funds provide diversified exposure to stocks.
The share market reacts to corporate earnings, inflation data, and central bank policies.
The Federal Reserve and Reserve Bank of India influence market sentiment through monetary policy.
Global events such as wars, pandemics, and elections affect stock market volatility.
"""

print("Total Words:", len(share_market_paragraph.split()))
```

```python
doc = nlp(share_market_paragraph)

entities = []
for ent in doc.ents:
    entities.append([ent.text, ent.label_])

# Convert to DataFrame
df_ner = pd.DataFrame(entities, columns=["Entity", "Label"])
entity_counts = Counter(df_ner["Label"])

print("\nEntity Type Counts:")
for label, count in entity_counts.items():
    print(f"{label}: {count}")
print("\nSample Named Entities:")
print(df_ner.head(25))
```
